```python
import pandas as pd
import json
import os
```

```python
path = r"C:\Users\Akshaya\Downloads\sample_airbnb.json"

Data=open(path,'r')
D=json.load(Data)
```

```python
stay_table={'Stay_id':[],'Stay_name':[],'Stay_description':[],'Summary':[],'Neighborhood_description':[],'Transit_detail':[],'Access_from_Stay':[],'House_rules':[],'Property_type':[],'Room_type':[],'Bed_type':[],'Minimum_Nights':[],'Maximum_Nights':[],'Cancellation_policy':[],'Can_accomodate':[],'Bedrooms':[],'Beds':[],'NumberOfReviews':[],'Bathroom':[],'Amenities':[],'Price':[],'Security_deposit':[],'Cleaning_fee':[],'Extra_people':[],'Guest_Included':[],'Stay_image':[],'Address':[],'Country':[],'Cordinates':[],'Availability_in_365':[],'Review_Accuracy':[],'Review_cleanliness':[],'Review_Location':[],'Review_communication':[],'Review_checkin':[],'Review_value':[],'Review_rating_100':[],'Last_review_added':[]}
for x in range (len(D)):
    stay_table['Stay_id'].append(D[x]['_id'])
    stay_table['Stay_name'].append(D[x]['name'])
    stay_table['Stay_description'].append(D[x]['description'])
    stay_table['Summary'].append(D[x]['summary'])
    stay_table['Neighborhood_description'].append(D[x]['neighborhood_overview'])
    stay_table['Transit_detail'].append(D[x]['transit'])
    stay_table['Access_from_Stay'].append(D[x]['access'])
    stay_table['House_rules'].append(D[x]['house_rules'])
    stay_table['Property_type'].append(D[x]['property_type'])
    stay_table['Room_type'].append(D[x]['room_type'])
    stay_table['Bed_type'].append(D[x]['bed_type'])
    stay_table['Minimum_Nights'].append(D[x]['minimum_nights'])
    stay_table['Maximum_Nights'].append(D[x]['maximum_nights'])
    stay_table['Cancellation_policy'].append(D[x]['cancellation_policy'])
    stay_table['Can_accomodate'].append(D[x]['accommodates'])
    try:
        stay_table['Bedrooms'].append(D[x]['bedrooms'])
    except:
        stay_table['Bedrooms'].append('Not_Provided')
    try:
        stay_table['Beds'].append(D[x]['beds'])
    except:
         stay_table['Beds'].append('Not_Provided')
    stay_table['NumberOfReviews'].append(D[x]['number_of_reviews'])
    try:
        stay_table['Bathroom'].append(D[x]['bathrooms'])
    except:
        stay_table['Bathroom'].append('Not_Provided')
    stay_table['Amenities'].append(D[x]['amenities'])
    stay_table['Price'].append(D[x]['price'])
    try:
        stay_table['Security_deposit'].append(D[x]['security_deposit'])
    except:
        stay_table['Security_deposit'].append(0)
    try:
        stay_table['Cleaning_fee'].append(D[x]['cleaning_fee'])
    except:
        stay_table['Cleaning_fee'].append(0)
    stay_table['Guest_Included'].append(D[x]['guests_included'])
    stay_table['Extra_people'].append(D[x]['extra_people'])
    stay_table['Stay_image'].append(D[x]['images']['picture_url'])
    stay_table['Address'].append(D[x]['address']['market'])
    stay_table['Country'].append(D[x]['address']['country'])
    stay_table['Cordinates'].append(D[x]['address']['location']['coordinates'])
    stay_table['Availability_in_365'].append(D[x]['availability']['availability_365'])
    try:
        stay_table['Last_review_added'].append(D[x]['last_review'])
    except:
        stay_table['Last_review_added'].append('Not_provided')
    try:
        stay_table['Review_Accuracy'].append(D[x]['review_scores']['review_scores_accuracy'])
    except:
        stay_table['Review_Accuracy'].append('No')
    try:
        stay_table['Review_checkin'].append(D[x]['review_scores']['review_scores_checkin'])
    except:
        stay_table['Review_checkin'].append('No')
    try:
        stay_table['Review_cleanliness'].append(D[x]['review_scores']['review_scores_cleanliness'])
    except:
        stay_table['Review_cleanliness'].append('No')
    try:
        stay_table['Review_communication'].append(D[x]['review_scores']['review_scores_communication'])
    except:
        stay_table['Review_communication'].append('No')
    try:
        stay_table['Review_Location'].append(D[x]['review_scores']['review_scores_location'])
    except:
        stay_table['Review_Location'].append('No')
    try:
        stay_table['Review_rating_100'].append(D[x]['review_scores']['review_scores_rating'])
    except:
        stay_table['Review_rating_100'].append('No')
    try:
        stay_table['Review_value'].append(D[x]['review_scores']['review_scores_value'])
    except:
        stay_table['Review_value'].append('No')

Stay_Table=pd.DataFrame(stay_table)
```

```python
host_table={'Stay_id':[],'Host_id':[],'Host_Name':[],'Host_location':[],'Host_about':[],'Host_superhost':[],'Host_identity_verified':[],'Host_listing':[],'Host_verification':[]}
for x in  range (len(D)):
    host_table['Stay_id'].append(D[x]['_id'])
    host_table['Host_id'].append(D[x]['host']['host_id'])
    host_table['Host_Name'].append(D[x]['host']['host_name'])
    host_table['Host_location'].append(D[x]['host']['host_location'])
    host_table['Host_about'].append(D[x]['host']['host_about'])
    host_table['Host_superhost'].append(D[x]['host']['host_is_superhost'])
    host_table['Host_identity_verified'].append(D[x]['host']['host_identity_verified'])
    host_table['Host_listing'].append(D[x]['host']['host_total_listings_count'])
    host_table['Host_verification'].append(D[x]['host']['host_verifications'])
Host_Table=pd.DataFrame(host_table)
```

```python
review_table={'Stay_id':[],'Reviewer_id':[],'Reviewer_name':[],'Date':[],'Comments':[]}
for x in range (len(D)):
    for y in range (len(D[x]['reviews'])):
        review_table['Stay_id'].append(D[x]['_id'])
        review_table['Reviewer_id'].append(D[x]['reviews'][y]['reviewer_id'])
        try:
            review_table['Reviewer_name'].append(D[x]['reviews'][y]['reviewer_name'])
        except:
            review_table['Reviewer_name'].append("Not Provided")
        review_table['Date'].append(D[x]['reviews'][y]['date'])
        try:
            review_table['Comments'].append(D[x]['reviews'][y]['comments'])
        except:
            review_table['Comments'].append('No Comments')
Review_Table=pd.DataFrame(review_table)
```
